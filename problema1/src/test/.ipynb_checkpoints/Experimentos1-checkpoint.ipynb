# Tiempos de ejecucion de Prim, para valores de n de 10 a 20, con grafos conexos, siendo m = O(nÂ²)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random

sns.set()

plt.clf()

dataframePrim = pd.read_csv("./tiemposPrim/tiemposN.csv")
tiemposPrim = []
vertices = []

# itero el rango de los valores de n considerados
for n in range(10, 51):
    tiemposPrim.append(dataframePrim[dataframePrim['vertices'] == n]['tiempo'].mean())
    vertices.append(n)

line1, = plt.plot(vertices, tiemposPrim, label="Prim")

plt.legend([line1], ['Prim'])
plt.xlabel('Vertices')
plt.ylabel('Tiempos en milisegundos')
plt.show()
```
