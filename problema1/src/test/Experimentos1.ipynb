# Tiempos de ejecucion de Prim, para valores de n variables, con grafos conexos, eligiendo m = O(nÂ²)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from math import pow

sns.set()

plt.clf()

dataframePrim = pd.read_csv("./complejidades/tiemposComplejidadPrim.csv")
tiemposPrim = []
vertices = []

min_n = dataframePrim['vertices'].min()
max_n = dataframePrim['vertices'].max()
# itero el rango de los valores de n considerados
for n in range(min_n, max_n+1):
    # Siendo m = O(n^2), el algoritmo es O(n^4). Entonces, dividirlo por
    # algo O(n^3) deberia resultar en una funcion similar a O(n).
    # currentNumber = n*(n-1)*(n-2)/2.
    currentNumber = pow(n, 3)
    tiemposPrim.append(dataframePrim[dataframePrim['vertices'] == n]['tiempo'].mean()/ currentNumber)
    vertices.append(n)

# z son los coeficientes del polinomio
z = np.polyfit(vertices, tiemposPrim, 1)
# p es un polinomio con los coeficientes de z
p = np.poly1d(z)

line1, = plt.plot(vertices, tiemposPrim, '.' ,label="Prim")
line2, = plt.plot(vertices, p(vertices))

plt.legend([line1, line2], ['Valores de Prim modificados', 'Ajuste por funciÃ³n lineal'])

# plt.savefig('./complejidadPrim.png')

plt.xlabel('Vertices')
plt.ylabel('Tiempos en milisegundos')
plt.show()
```

# Tiempos de ejecucion de Kruskal, para valores de n variables, con grafos conexos, eligiendo m = O(n^2)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from math import pow

sns.set()

plt.clf()

dataframeKruskal = pd.read_csv("./complejidades/tiemposComplejidadKruskal.csv")
tiemposKruskal = []
vertices = []

min_n = dataframeKruskal['vertices'].min()
max_n = dataframeKruskal['vertices'].max()
# itero el rango de los valores de n considerados
for n in range(min_n, max_n+1):
    # Siendo m = O(n^2), el algoritmo es O(n^5). Entonces, dividirlo por
    # algo O(n^4) deberia resultar en una funcion similar a O(n).
    # currentNumber = (n-1)*(n-2)*(n-1)*(n-2)/4.
    currentNumber = pow(n, 4)
    tiemposKruskal.append(dataframeKruskal[dataframeKruskal['vertices'] == n]['tiempo'].mean()/ currentNumber)
    vertices.append(n)

# z son los coeficientes del polinomio
z = np.polyfit(vertices, tiemposKruskal, 1)
# p es un polinomio con los coeficientes de z
p = np.poly1d(z)

line1, = plt.plot(vertices, tiemposKruskal, '.' ,label="Kruskal")
line2, = plt.plot(vertices, p(vertices))

plt.legend([line1, line2], ['Valores de Kruskal modificados', 'Ajuste por funciÃ³n lineal'])

# plt.savefig('./complejidadKruskal.png')

plt.xlabel('Vertices')
plt.ylabel('Tiempos en milisegundos')
plt.show()
```

# Tiempos de Kruskal con Path Compression, para valores de n variables, con grafos conexos, eligiendo m = O(n^2)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from math import pow
from math import log

sns.set()

plt.clf()

dataframeKruskalCompression = pd.read_csv("./complejidades/tiemposComplejidadKruskalCompression.csv")
tiemposKruskalCompression = []
vertices = []

min_n = dataframeKruskalCompression['vertices'].min()
max_n = dataframeKruskalCompression['vertices'].max()
# itero el rango de los valores de n considerados
for n in range(min_n, max_n+1):
    # Siendo m = O(n^2), el algoritmo es O(n^4*log(n)). Entonces, dividirlo por
    # algo O(n^3*log(n)) deberia resultar en una funcion similar a O(n).
    currentNumber = log(n, 2)*pow(n, 3)
    tiemposKruskalCompression.append(dataframeKruskalCompression[dataframeKruskalCompression['vertices'] == n]['tiempo'].mean()/ currentNumber)
    vertices.append(n)

# z son los coeficientes del polinomio
z = np.polyfit(vertices, tiemposKruskalCompression, 1)
# p es un polinomio con los coeficientes de z
p = np.poly1d(z)

line1, = plt.plot(vertices, tiemposKruskalCompression, '.' ,label="KruskalPathCompression")
line2, = plt.plot(vertices, p(vertices))

plt.legend([line1, line2], ['Valores de Kruskal con P-C modificados', 'Ajuste por funciÃ³n lineal'])

# plt.savefig('./complejidadKruskalCompression.png')

plt.xlabel('Vertices')
plt.ylabel('Tiempos en milisegundos')
plt.show()
```

# ComparaciÃ³n del caso promedio entre Prim y Kruskal, para n entre 10 y 70

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from math import pow
from math import log

sns.set()

plt.clf()

dataframePrimKruskal = pd.read_csv("./casosPromedio/promedioPrimKruskal.csv")

min_n = dataframePrimKruskal['vertices'].min()
max_n = dataframePrimKruskal['vertices'].max()

vertices = []
tiemposPrim = []
tiemposKruskal = []

for n in range(min_n, max_n+1):
    vertices.append(n)
    tiemposPrim.append(dataframePrimKruskal[dataframePrimKruskal['vertices'] == n]['tiempoPrim'].mean())
    tiemposKruskal.append(dataframePrimKruskal[dataframePrimKruskal['vertices'] == n]['tiempoKruskal'].mean())

line1, = plt.plot(vertices, tiemposPrim, label="Prim")
line2, = plt.plot(vertices, tiemposKruskal, label="Kruskal")

plt.legend([line1, line2], ['Prim', 'Kruskal'])
plt.xlabel('Vertices')
plt.ylabel('Tiempos en milisegundos')
# plt.savefig('./promedioPrimKruskal.png')
plt.show()
```
