```python
# coding=utf-8
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from math import pow

sns.set()

plt.clf()

dataframe = pd.read_csv("./tiemposN.csv")
tiempos = []
vertices = []

min_n = dataframe['vertices'].min()
max_n = dataframe['vertices'].max()

# itero el rango de los valores de n considerados
for n in range(min_n, max_n+1):  
    # Siendo m = O(n^2), el algoritmo es O(n^3). Entonces, dividirlo por
    # algo O(n^2) deberia resultar en una funcion similar a O(n).
    currentNumber = pow(n, 2)
    tiempos.append(dataframe[dataframe['vertices'] == n]['tiempo'].mean()/currentNumber)
    vertices.append(n)

# z son los coeficientes del polinomio
z = np.polyfit(vertices, tiempos, 1)
# p es un polinomio con los coeficientes de z
p = np.poly1d(z)

line1, = plt.plot(vertices, tiempos, '.', label="Floyd-Warshall")
line2, = plt.plot(vertices, p(vertices))

plt.legend([line1, line2], ['Valores de Floyd-Warshall modificados', 'Ajuste por funcion lineal'])

plt.savefig('./tiemposFloydWarshall.png')

plt.xlabel('Vertices')
plt.ylabel('Tiempos en milisegundos/f')
plt.show()
```
