# Compraracion entre Floyd-Warshall y Dantzig

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from math import pow
from math import log

sns.set()

plt.clf()

dataframeFloydWarshall = pd.read_csv("./tiemposFloydWarshall/tiemposN.csv")
dataframeDantzig = pd.read_csv("./tiemposDantzig/tiemposN.csv")

min_n = dataframeFloydWarshall['vertices'].min()
max_n = dataframeFloydWarshall['vertices'].max()

vertices = []
tiemposFloydWarshall = []
tiemposDantzig = []

escalaToSegundos = 1000

for n in range(min_n, max_n+1):
    vertices.append(n)
    tiemposFloydWarshall.append(dataframeFloydWarshall[dataframeFloydWarshall['vertices'] == n]['tiempo'].mean()/escalaToSegundos)
    tiemposDantzig.append(dataframeDantzig[dataframeDantzig['vertices'] == n]['tiempo'].mean()/escalaToSegundos)
    
line1, = plt.plot(vertices, tiemposFloydWarshall, label="Floyd-Warshall")
line2, = plt.plot(vertices, tiemposDantzig, label="Dantzig")

plt.legend([line1, line2], ['Floyd-Warshall', 'Dantzig'])
plt.xlabel('Vertices')
plt.ylabel('Tiempos en segundos')
plt.savefig('./comparacionFloydWarshallVsDantzig.png')
plt.show()
```
